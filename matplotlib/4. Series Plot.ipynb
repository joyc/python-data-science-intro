```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
s = Series(np.random.randn(1000))
s.plot(kind='kde', color='r')
```

```python
s1 = Series([1,2,3,4,5])
s1.cumsum()
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
s1.plot(kind='line', label='s1')
s2.plot(kind='line', label='s2')
plt.legend()
```

```python
fig, ax = plt.subplots(2, 1)
s1.plot(ax=ax[0], color='r', label='s1', title='s1 title')
ax[0].legend()
s2.plot(ax=ax[1], color='g', label='s2', title='s2 title')
ax[1].legend()
plt.tight_layout()
```
