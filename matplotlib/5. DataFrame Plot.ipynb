```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
df = DataFrame(np.random.randint(1, 10, 40).reshape(10, 4), columns=['A','B','C','D'])
df
```

```python
df.plot() #默认按照columns画图
```

```python
# 按照index画图
for i in df.index:
    df.loc[i].plot(label=str(i))
plt.legend()
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='barh') #barh 横过来
```

```python
df.plot(kind='bar', stacked=True) # 叠加柱形图
```

```python
df.plot(kind='area') # 面积图
```

```python
df['A'].plot() # 单列也可以画图 因为是 Series
```
