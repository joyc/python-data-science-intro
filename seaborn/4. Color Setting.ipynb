```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# 定义显示函数
def sinplot():
    x = np.linspace(0,14,100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='sin(x+1)*(i+0.75)')
    plt.legend()
```

```python
sinplot()
```

```python
import seaborn as sns
```

```python
sns.set()
sinplot()
```

```python
# 使用with来自动恢复
with sns.axes_style('whitegrid'):
    sinplot()
```

```python
with sns.color_palette('bright'):
    sinplot()
```
