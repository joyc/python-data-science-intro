```python
# basic
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

# matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# seaborn
import seaborn as sns
```

```python
s = Series(np.random.randn(1000))
```

```python
s.plot(kind='hist')
```

```python
plt.hist(s)
```

```python
s.plot(kind='kde')
```

# 通过Seaborn作图

```python
sns.distplot(s)
```

```python
sns.distplot(s, hist=False, color='r')
```

```python
sns.distplot(s, kde=False, bins=30)
```

```python
sns.kdeplot(s, color='r', shade=True)
```
