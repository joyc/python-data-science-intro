```python
# basic
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

# matplotlib
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 14, 100) # 生成数列
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
# 定义显示plot的函数
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
sinplot()
```
