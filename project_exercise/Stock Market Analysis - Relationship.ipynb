```python
# basic
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

# get data
import pandas_datareader as pdr

# visual
import matplotlib.pyplot as plt
import seaborn as sns

# time
from datetime import datetime

# set seaborn
sns.set()
```

```python
df = pdr.get_data_yahoo('BABA')
df.head()
```

```python
start = datetime(2016,1,1)
company = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'FB']
```

```python
top_df = pdr.get_data_yahoo(company, start=start)
```

```python
type(top_df)
```

```python
top_df = top_df['Adj Close']
top_df.head()
```

```python
top_df_dr = top_df.pct_change()
```

```python
top_df.plot(figsize=(10,5))
```

```python
top_df_dr.plot()
```

```python
sns.jointplot('AMZN', 'GOOG', top_df_dr, kind='scatter') # 散点图
```

```python
sns.pairplot(top_df_dr.dropna())
```
